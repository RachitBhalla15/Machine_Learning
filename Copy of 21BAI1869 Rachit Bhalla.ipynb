#Fill the ST_NUM by the maximum number of houses in a particular street

```python
#
propd['ST_NUM'] = propd.groupby('ST_NAME')['ST_NUM'].transform(lambda x: x.fillna(x.max()))
print(propd['ST_NUM'].isna().sum())
```
